```python
import pandas as pd
import datetime
import yfinance as yf

def fetch_closing_prices(symbol, start_date, end_date):
    try:
        # Create a Yahoo Finance Ticker object for the company symbol
        ticker = yf.Ticker(symbol)

        # Fetch historical data from Yahoo Finance
        historical_data = ticker.history(start=start_date, end=end_date)

        # Extract the 'Close' column and reset the index
        closing_prices = historical_data[['Close']].reset_index()

        return closing_prices

    except Exception as e:
        print(f"Error fetching data for {symbol}: {e}")
        return None

# Define the new date range
start_date_new = datetime.date(2013, 7, 1)
end_date_new = datetime.date(2023, 10, 9)

# Create an empty DataFrame to store closing prices
closing_prices_df = pd.DataFrame()

sector_abbreviations = {
    'Automobile': 'ATO',
    'Banking': 'BNK',
    'Cement': 'CMT',
    'Cigarettes & FMCG': 'C&F',
    'Conglomerate': 'CGM',
    'Diamond & Jewellery': 'D&J',
    'Engineering & Construction': 'E&C',
    'Finance': 'FN',
    'FMCG': 'FMCG',
    'IT Services & Consulting': 'IT',
    'Oil exploration and Production': 'OEP',
    'Paints': 'PNTS',
    'Pharmaceuticals': 'PHR',
    'Power generation/Distribution': 'PWR',
    'Steel': 'STL',
    'Telecommunication': 'TLE'
}

# Fetch closing prices for each additional company and add them as columns to the DataFrame
for company in additional_bse_companies:
    sector = company["sector"]
    symbol = company["symbol"]
    abbreviation = sector_abbreviations.get(sector, "UNK")  # Get sector abbreviation or use "UNK" for unknown sectors

    # Fetch historical closing prices for the company
    company_data = fetch_closing_prices(symbol, start_date_new, end_date_new)

    if company_data is not None:
        # Rename the 'Close' column to the company's symbol with sector abbreviation
        column_name = f"{abbreviation}_{symbol[:-3]}"  # Remove '.BO' from the symbol
        company_data.rename(columns={"Close": column_name}, inplace=True)

        # Merge the data into the main DataFrame based on the 'Date' column
        if closing_prices_df.empty:
            closing_prices_df = company_data
        else:
            closing_prices_df = closing_prices_df.merge(company_data, on="Date", how="outer")

# Fill missing values with NaN or any other appropriate value
closing_prices_df.fillna(method="ffill", inplace=True)

closing_prices_df['Date'] = closing_prices_df['Date'].dt.strftime('%Y-%m-%d')

closing_prices_df.sort_values(by="Date", inplace=True)

# Save the DataFrame to a CSV file
closing_prices_df.to_csv("bse_closing_price_new.csv", index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.optimizers import Adam
df = pd.read_csv("new_csv.csv")
df["Date"] = pd.to_datetime(df["Date"])
start_date = pd.to_datetime("2013-07-01")
end_date = pd.to_datetime("2023-10-09")
df = df[(df["Date"] >= start_date) & (df["Date"] <= end_date)]
closing_prices = df["E&C_SADBHAV"].values.reshape(-1, 1)
scaler = MinMaxScaler()
closing_prices = scaler.fit_transform(closing_prices)
train_size = int(len(closing_prices) * 0.8)
train, test = closing_prices[:train_size], closing_prices[train_size:]
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        seq = data[i:i + seq_length]
        target = data[i + seq_length]
        X.append(seq)
        y.append(target)
    return np.array(X), np.array(y)
seq_length = 10 
X_train, y_train = create_sequences(train, seq_length)
X_test, y_test = create_sequences(test, seq_length)
model = Sequential()
model.add(LSTM(50, input_shape=(X_train.shape[1], 1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer=Adam(learning_rate=0.001))
model.fit(X_train, y_train, epochs=50, batch_size=64, verbose=1)
y_pred = model.predict(X_test)
y_pred = scaler.inverse_transform(y_pred)
y_test = scaler.inverse_transform(y_test)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
print(f"R-squared: {r2}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
plt.plot(y_test, label='Actual Price', color='black')
plt.plot(y_pred, label='Predicted Price', color='green')
plt.xlabel('Time')
plt.ylabel('Closing Price')
plt.title('LSTM Predicted vs Actual Closing Price')
plt.legend()
plt.show()
```
